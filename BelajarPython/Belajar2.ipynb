```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('classic')
```

```python
#definisi fungsi 
def returns_dydt(y, t):
    Nps, Npt, Nds, Ndt = y
    dydt = [(-l1 * Nps - k * Nps) , (k * Nps - l2 * Npt), (l1 * Nps - k * Nds), (l2 * Npt + k * Nds)]
    return dydt
```

```python
#insiasi 
y0 = [10,0,0,0]
l1 = 0.5
l2 = 0.5
k = 0.8
```

```python
#fungsi waktu
t = np.linspace(0,10,100)
```

```python
#hasil fungsi 
y = odeint(returns_dydt, y0, t)
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(t, y[:,0], label = 'Nps')
plt.plot(t, y[:,1], label = 'Npt')
plt.plot(t, y[:,2], label = 'Nds')
plt.plot(t, y[:,3], label = 'Ndt')
plt.xlabel("Time")
plt.legend(loc='best')
plt.grid()
plt.show()
```
